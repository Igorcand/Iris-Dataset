```python
from sklearn.datasets import load_iris
iris = load_iris()
x = iris.data  # A lista de valores
y = iris.target  #O resultado 






```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
#print(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
valores_performance = {}
for num in range(1, 26):
    knn = KNeighborsClassifier(n_neighbors=num)
    knn.fit(x_train, y_train)
    previsao = knn.predict(x_test)
    acertos = metrics.accuracy_score(y_test, previsao)
    valores_performance[num] = round(acertos, 4)
    #print(f'Neighbors {num} -- Porcentagem {acertos:.5f}')
    
print(valores_performance)
    
    
```

```python
import mathplotlib.pyplot as plt

plt.plot(list(valores_performance.keys()), list(valores_performance.values()))
plt.xlabel('Valores de K')
plt.ylabel('Performance')  

plt.show()
```
