# Iris Dataset 

### Importação dos dados

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

### Observações

```python
x = iris.data

print(iris.data)  #cada um das linhas é uma amostra, então temos 150 amostras, compri_petala e larg_petada / compri_sepala e larg_sepala
#Isso parece uma lista  mas é um array numpy, muito usado em machine learning
```

### Target / Resultados


```python
y = iris.target
print(iris.target) # no caso desse exemplo, o target são as espécies da planta
#para saber o nome das especies das plantas so usar o target_name
```

### Shapes das observações

```python
print(iris.data.shape)  #Se tivessemos usando uma lista normal, poderiamos usar o len(), mas como é um array numpy não funciona
#no caso temos 150 observações(resultados) e 4 atributos(featcher)
```

### Shapes dos target

```python
print(iris.target.shape) # apenas o numero de resultados
```

# Importação do KNN

#### O KNN é um modelo estatístico que vamos escolher para testar a performance
#### KNN (K Nearest neighbors) K vizinhos mais próximos.

```python
from sklearn.neighbors import KNeighborsClassifier  #Importação da biblioteca
knn = KNeighborsClassifier(n_neighbors=1) #criamos uma variável para utilizar o KNN e definimos o valor de K vizinhos = 1
```

## Treinar a máquina

```python
knn.fit(x, y)  # Rodado uma vez ele vai aprender e fazer os calculos de acordo com as informações já informadas pra ele
```

## Fazer previsões

```python
species = knn.predict([[5.9, 3,5.1, 1.8]])[0] # Ele vai fazer a previsão de acordo com o que ele aprendeu em knn.fit() 
print(iris.target_names[species])  #mostra o resultado da previsão dele
```

### Separar os dados em dois grupos

```python
from sklearn.model_selection import train_test_split  #Precisamos de 2 variaveis x(iris.data) e 2 variaveis y(iris.target)
x_train, x_test, y_train,y_test = train_test_split(x, y, test_size = 0.25) #Uma (x, y) para treinar e outra para testar
print(x_test.shape) #Vemos quantas foram usadas para teste              test_size= 0.25 ->é a porcentagem queele vai usar para treinar 
```

### Avaliação da performance do modelo

```python
knn.fit(x_train, y_train) #Ele vai aprender de acordo com os 25% definidos acima para o x e y de teste
previsoes = knn.predict(x_test) #previsoes feitas a partir do aprendizado do x e y de teste
print(previsoes)
```

```python
from sklearn import metrics  #calcular porcentagem de acertos
acertos = metrics.accuracy_score(y_test, previsoes)
print(acertos)
```

## Aplicação do modelo de regressão logística

```python
#É um  modelo de previsão logica, mais usada com a classificação binária
#exemplos:
#Se um email é ou não spam
#se um tumor é maligno ou benigno
#Mas pode ser usada tambem como mais de 2 parametros
```

##### Como ela preferêncialmente é usada com classificações binária, o eixo y dela é dividido em 0 e 1, com uma função que é calculada, vai definir uma curva em formato de S entre elas, e assim, de acordo com o valor do eixo X verifica uma certa porcentagem de ela ser ou não de um grupo. 

#### No caso das espécies de planta, que são 3 espécies, o algorítimo faz essa mesma conta pelo gráfico, e faz 3 gráficos na seguinte lógica, na porcentagem de ela ser(ser da espécie) ou não ser(2 espécies) da espécie. Então comparando o numero nos 3 gráficos das espécies de planta ele consegue achar a espécie.

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(x_train, y_train)
previsoes_logreg = logreg.predict(x_test)
acertos_logreg = metrics.accuracy_score(y_test, previsoes_logreg)
print(acertos_logreg)
```
